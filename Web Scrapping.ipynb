```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
Product_name = []
Prices = []
Description = []
Reviews = []
```

```python
url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"+str(2)
r = requests.get(url)
soup = BeautifulSoup(r.text,"lxml")
```

```python
names = soup.find_all("div",class_ = "_4rR01T")
```

```python
for i in names:
    name = i.text
    Product_name.append(name) 
print(Product_name)
```

```python
descriptions = soup.find_all("div",class_ = "fMghEO")
for i in descriptions:
    description = i.text
    Description.append(description) 
print(Description)
```

```python
prices = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")
```

```python
for i in prices:
    price = i.text
    Prices.append(price)
print(Prices)
```

```python
reviews = soup.find_all("div", class_ = "gUuXy-")
for i in reviews:
    review = i.text
    Reviews.append(review)
print(Reviews)
```

```python
df = pd.DataFrame({"Product name": Product_name, "Prices":Prices, "Discription":Description,"Reviews":Reviews})
```
