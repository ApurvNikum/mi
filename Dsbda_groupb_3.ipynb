```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\priyanshu borole\\Downloads\\airquality.csv")
```

```python
df
```

```python
df = df.drop("Unnamed: 0",axis=1)
df
```

```python
df.isnull().sum()
```

```python
df["Ozone"] = df['Ozone'].fillna(df['Ozone'].mean())
df["Solar.R"] = df['Solar.R'].fillna(df['Solar.R'].mean())
df["Humidity"] = df['Humidity'].fillna(df['Humidity'].mode()[0])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['Humidity']=label.fit_transform(df['Humidity'])
df['Humidity'].unique()
```

```python
df
```

```python
plt.scatter(x=df['Ozone'],y=df['Temp'])
plt.legend(['Ozone'])
plt.xlabel("ozone")
plt.ylabel("label")
plt.title("Ozone Scatter Plot")
plt.show()
```

```python
sns.barplot(x = df['Month'],y = df['Temp'])
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
```

```python
sns.pairplot(df)
```

```python
h = df.loc[1:15,['Ozone']]
v = df.loc[1:15,['Temp']]
plt.plot(h,label = 'Ozone',marker = 'o', linestyle = "dotted")
plt.plot(v,label = 'Temp',marker = 'o', linestyle = "dotted")
plt.title("Line Graph")
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df['Month'],y=df['Temp'])
```

```python
h = df.loc[:,'Humidity']
plt.hist(h,bins='auto')
plt.title("Histogram")
plt.xlabel("Humidity")
plt.show()
```
